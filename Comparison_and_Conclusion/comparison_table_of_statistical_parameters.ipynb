- Standard Cubic Foot, or SCF for short, is the volume of gas at standard circumstances of pressure and temperature, which are usually expressed as 14.7 psi (1 atmosphere) and 60°F (15.6°C).

- A 42-gallon barrel of oil tested under normal conditions is referred to as an STB (stock tank barrel). This is usually done after the oil has been stabilized by eliminating any dissolved gasses and separating any water content.

- Consequently, SCF/STB measures the volume of gas that, under typical circumstances, can dissolve in a single barrel of oil from a stock tank.

```python
import matplotlib.pyplot as plt

# Data
pressure = [2405, 2200, 1950, 1700, 1450, 1200, 950, 700, 450, 200]
experimental = [737, 684, 620, 555, 492, 429, 365, 301, 235, 155]
glaso = [665, 500, 425, 388, 319, 285, 205, 165, 100, 55]
standing = [703, 735, 658, 575, 493, 309, 242, 185, 100, 45]
petrosky_fashad = [685, 508, 467, 398, 342, 297, 255, 200, 170, 125]
ant_co = [734, 682, 619, 554, 492, 429, 366, 302, 237, 157]

# Plot
plt.figure(figsize=(6, 6))
plt.plot(pressure, experimental, 'o-', color='blue', label='Experimental')
plt.plot(pressure, glaso, 's--', color='orange', label='Glaso')
plt.plot(pressure, standing, '^-.', color='green', label='Standing')
plt.plot(pressure, petrosky_fashad, 'd:', color='red', label='Petrosky and Fashad')
plt.plot(pressure, ant_co, 'x-', color='purple', label='Ant Colony')

# Title and labels
plt.title('Comparison of Gas Solubility Models')
plt.xlabel('Pressure (psig)')
plt.ylabel('Gas Solubility (scf/stb)')
plt.legend()
plt.grid(True)
plt.show()
```
